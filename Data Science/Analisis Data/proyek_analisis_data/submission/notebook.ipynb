## Data Wrangling

### Gathering Data

Data akan dikumpulkan dari file csv yang ada pada dataset Air Quality yang ada. 

Berhubung data tiap stasiun dipisahkan menjadi file csv yang berbeda, dengan total 12 file csv data Air Quality di tiap statiun. Proses iterasi akan dilakukan untuk menggabungkan data satu per satu

```python
customer_df = pd.read_csv('./E-Commerce Public Dataset/customers_dataset.csv')
order_df = pd.read_csv('./E-Commerce Public Dataset/orders_dataset.csv')
order_item_df = pd.read_csv('./E-Commerce Public Dataset/order_items_dataset.csv')
order_payment_df = pd.read_csv('./E-Commerce Public Dataset/order_payments_dataset.csv')
order_review_df = pd.read_csv('./E-Commerce Public Dataset/order_reviews_dataset.csv')
products_df = pd.read_csv('./E-Commerce Public Dataset/products_dataset.csv')
products_category_df = pd.read_csv('./E-Commerce Public Dataset/product_category_name_translation.csv')
```

```python
customer_df.head(3)
```

```python
customer_df.drop(columns=['customer_unique_id', 'customer_zip_code_prefix'], axis=1, inplace=True)
```

```python
order_df.head(3)
```

```python
order_item_df.head(3)
```

```python
order_item_df.drop(columns=['seller_id', 'shipping_limit_date'], axis=1, inplace=True)
```

```python
order_payment_df.head(3)
```

```python
products_df.head(3)
```

```python
products_df = products_df[['product_id', 'product_category_name']]
```

```python
products_category_df.head(3)
```

#### Join

```python
new_order_df = pd.merge(
    left=order_df,
    right=customer_df,
    how="inner",
    left_on="customer_id",
    right_on="customer_id"
)
```

```python
new_order_df.head()
```

```python
order_item = pd.merge(
    left=new_order_df,
    right=order_item_df,
    how="inner",
    left_on="order_id",
    right_on="order_id"
)
```

```python
order_item.head()
```

```python
order_item_payment = pd.merge(
    left=order_item,
    right=order_payment_df,
    how="inner",
    left_on="order_id",
    right_on="order_id"
)
```

```python
order_item_payment.head(3)
```

```python
product_category = pd.merge(
    left=products_df,
    right=products_category_df,
    how="inner",
    left_on="product_category_name",
    right_on="product_category_name"
)
```

```python
product_category.head()
```

```python
all_df = pd.merge(
    left=order_item_payment,
    right=product_category,
    how="inner",
    left_on="product_id",
    right_on="product_id"
)
```

```python
all_df.head()
```
